```python
from selenium import webdriver
from selenium.webdriver.safari.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time

def getCal(month):
    
    # Initialize the Safari WebDriver
    service = Service() 
    driver = webdriver.Safari(service=service)
    
    # URL to scrape
    url = "https://cbase.som.sunysb.edu/conet/calendar/dynamicCalendar.cfm"
    #url='https://cbase.som.sunysb.edu/conet/calendar/stud_mainFrame.cfm'
    
    # Open a page to set cookies
    driver.get("https://cbase.som.sunysb.edu/conet/calendar/calendarFrame.cfm")
    
    # Clear all cookies before setting new ones
    driver.delete_all_cookies()
    
    # Set cookies
    cookies = [
        {'name': 'CFID', 'value': '14209121'},
        {'name': 'CFTOKEN', 'value': '36158840'}
    ]
    for cookie in cookies:
        driver.add_cookie(cookie)
    
    # Navigate to the target URL
    driver.get(url)

    # Wait for the content to load
    WebDriverWait(driver, 30).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, 'body'))
    )

    # Set month
    driver.execute_script(f"document.getElementsByName('tbSelMonth')[0].value = '{month}';")
    
    # Trigger the change event if needed
    driver.execute_script("document.getElementsByName('tbSelMonth')[0].dispatchEvent(new Event('change'));")
    time.sleep(2)
    
    # Extract month and year
    month_select_element = driver.find_element(By.NAME, 'tbSelMonth')
    month_selected_option = month_select_element.find_element(By.CSS_SELECTOR, 'option:checked')
    month_val = month_selected_option.get_attribute('value')
    
    year_select_element = driver.find_element(By.NAME, 'tbSelYear')
    year_selected_option = year_select_element.find_element(By.CSS_SELECTOR, 'option:checked')
    year_val = year_selected_option.get_attribute('value')
    
    # Extract HTML content
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    driver.quit()

    return soup
```

```python
import re
from datetime import datetime
import pytz

def createEventList(soup, month, year):

    # Month map
    months = {
        "1": "January", "2": "February", "3": "March", "4": "April",
        "5": "May", "6": "June", "7": "July", "8": "August",
        "9": "September", "10": "October", "11": "November", "12": "December"
    }
    timezone = pytz.timezone('America/New_York')
        
    # Get table (month)
    table_body = soup.select_one('body > form > table > tbody > tr:nth-child(2) > td > table > tbody')
    events = []

    # Iterate over days
    for cell in table_body.find_all('td', {'id': 'calCell'}):
        date = cell.find('font', {'id': 'calDateText'})
        if date:
            day = date.get_text(strip=True)

            # Iterate over classes
            for section in cell.find_all('table'):
                rows = section.find_all('tr')
                prev_event=None
                for row in rows:
                    if not prev_event:
                        prev_event = row.find('font', {'color': '#0033FF'})
                    else:
                        event_time = row.find('font', {'color': '#000000'})
                        event_name = prev_event.get_text(strip=True)
                        class_name = re.sub(r'^\*.\s*', '', event_name)
                        event_time = event_time.get_text(strip=True)

                        # Create date string
                        date_str = f"{day}-{month}-{year}"
                        date_format = "%d-%m-%Y"
                        date_obj = datetime.strptime(date_str, date_format)
                        
                        # Parse time range
                        start_time_str, end_time_str = event_time.split('-')
                        start_time_obj = datetime.strptime(start_time_str.strip(), "%I:%M %p")
                        end_time_obj = datetime.strptime(end_time_str.strip(), "%I:%M %p")
                        
                        # Combine date with times
                        start_datetime = datetime.combine(date_obj.date(), start_time_obj.time())
                        end_datetime = datetime.combine(date_obj.date(), end_time_obj.time())
                        start_datetime = timezone.localize(start_datetime)
                        end_datetime = timezone.localize(end_datetime)


                        # Create event dict
                        events.append({
                            'start_datetime': start_datetime,
                            'end_datetime': end_datetime,
                            'name': class_name,
                        })
                        prev_event=None
    return events
```

```python
from ics import Calendar, Event
from datetime import datetime 

def eventToCal(eventList):

    calendar = Calendar()
    
    for event in eventList:
        e = Event()
        e.name = event['name']
        e.begin = event['start_datetime']
        e.end = event['end_datetime']
        calendar.events.add(e)
    
    with open('events.ics', 'w') as f:
        f.writelines(calendar)
```

```python
def main():
    months = ["8", "9", "10", "11", "12"]
    eventList=[]
    for month in months:
        soup = getCal(month)
        test = createEventList(soup, month, '2024')
        eventList.extend(test)
    eventToCal(eventList)
```

```python
main()
```
